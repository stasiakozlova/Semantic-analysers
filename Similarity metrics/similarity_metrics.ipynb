```python
nltk.download('wordnet')
import nltk
from nltk.corpus import wordnet
```

```python
print(f'Path length: {syn1.path_similarity(syn2)}')
print(f'Leacock & Chodorow: {syn1.lch_similarity(syn2)}') 
print(f'Wu & Palmer: {syn1.wup_similarity(syn2)}')
```
